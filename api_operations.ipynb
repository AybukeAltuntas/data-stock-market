```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import requests

url = "https://api.polygon.io/v2/aggs/ticker/AAPL/range/1/day/2025-11-13/2026-02-11?adjusted=true&apiKey=ol3pGbgwovvTp0B6frXDjT_Bm7AxatKe"    # Daha önce bulduğunuz URL ile değiştirin
api_data = requests.get(url).json()
```

```python
api_data
```

```python
apple_df = pd.DataFrame(api_data["results"])

apple_df.head()
```

```python
apple_df["date"] = pd.to_datetime(apple_df["t"], unit="ms")
apple_df
```

```python
apple_df = apple_df.set_index("date")
apple_df
```

```python
mapping = { 
    'o': 'open', 
    'c': 'close', 
    'h': 'high', 
    'l': 'low', 
    'n': 'number', 
    'v': 'volume', 
    'vw':'avg_price' 
}
apple_df = apple_df.rename(columns=mapping)
apple_df
```

```python
apple_df["close"].plot()
```
